```python
import networkx as nx
import matplotlib.pyplot as plt
from scipy.spatial.distance import hamming
```

# Facebook-Ego

```python
G_fb = nx.read_edgelist("Group7/Facebook-Ego/1912.edges", create_using = nx.Graph(), nodetype = int)
```

```python
# 1 Calculate betweenness, closeness and PageRank for each node in the network
betweenness_centrality = nx.betweenness_centrality(G_fb)
print(betweenness_centrality)

```

```python
closeness_centrality = nx.closeness_centrality(G_fb)
print(closeness_centrality)
```

```python
pagerank_fb = nx.pagerank(G_fb)
print(pagerank_fb)
```

```python
# to compare  between betweenness, closeness and PageRank, 
# we rank the nodes based on each metric and print the 5top nodes in each ranking
betweenness_rank = sorted(betweenness_centrality, key=betweenness_centrality.get, reverse=True)
closeness_rank = sorted(closeness_centrality, key=closeness_centrality.get, reverse=True)
pagerank_rank = sorted(pagerank_fb, key=pagerank_fb.get, reverse=True)


print("Top Nodes based on Betweenness Centrality:")
print(betweenness_rank[:5])  

print("\nTop Nodes based on Closeness Centrality:")
print(closeness_rank[:5])  

print("\nTop Nodes based on PageRank:")
print(pagerank_rank[:5])  
```

The differences in results arise because each measure evaluates nodes based on different criteria:

    Betweenness Centrality: Considers nodes that act as crucial bridges in the network.
    Closeness Centrality: Favors nodes that can reach others quickly on average.
    PageRank: Highlights nodes that receive many important connections from other significant nodes.

These diverse criteria lead to variations in the identified top nodes for each measure.

# largest strong connected component in the Twitter-Ego dataset

```python
# plot the largest strong connected component in the Twitter-Ego dataset
G_tw = nx.read_edgelist("Group7/Twitter-Ego/2841151.edges", create_using = nx.DiGraph(), nodetype = int)
largest_scc = max(nx.strongly_connected_components(G_tw), key=len)
print("Number of nodes in the largest strongly connected component:", len(largest_scc))
scc_subgraph = G_tw.subgraph(largest_scc)

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(scc_subgraph, seed=42)  
nx.draw(scc_subgraph, pos, node_size=10)
plt.title("Largest Strongly Connected Component in Twitter-Ego Dataset")
```

```python
# next results are based on the facebook ego network
# Find the node with the highest PageRank value and the node with the lowest PageRank value
highest_pagerank_node = max(pagerank_fb, key=pagerank_fb.get)
print("highest_pagerank_node:", highest_pagerank_node)
lowest_pagerank_node = min(pagerank_fb, key=pagerank_fb.get)
print("lowest_pagerank_node:", lowest_pagerank_node)

jaccard_values_highest = list(nx.jaccard_coefficient(G_fb, [(highest_pagerank_node, n) for n in G_fb.nodes() if n != highest_pagerank_node]))
jaccard_values_lowest = list(nx.jaccard_coefficient(G_fb, [(lowest_pagerank_node, n) for n in G_fb.nodes() if n != lowest_pagerank_node]))
jaccard_values = [value[2] for value in jaccard_values_highest]
jaccard_values_lowest = [value[2] for value in jaccard_values_lowest]

```

```python
# Calculate the cosine similarity for the given nodes with the rest of the nodes
def cosine_similarity_nodes(graph, node1, node2):
    neighbors1 = set(graph.neighbors(node1))
    neighbors2 = set(graph.neighbors(node2))
    common_neighbors = neighbors1 & neighbors2
    degree_product = graph.degree[node1] * graph.degree[node2]
    if degree_product == 0:
        return 0
    else:
        return len(common_neighbors) / degree_product ** 0.5


cosine_similarities_highest = [cosine_similarity_nodes(G_fb, highest_pagerank_node, node) for node in G_fb.nodes() if node != highest_pagerank_node]
cosine_similarities_lowest = [cosine_similarity_nodes(G_fb, lowest_pagerank_node, node) for node in G_fb.nodes() if node != lowest_pagerank_node]
```

```python
# Calculate the Hamming distances
adj_matrix = nx.adjacency_matrix(G_fb).todense()
node_to_index = {node: index for index, node in enumerate(G_fb.nodes())}


hamming_distances_highest = [hamming(adj_matrix[node_to_index[highest_pagerank_node], :], adj_matrix[:,node_to_index[node]]) for node in G_fb.nodes() if node != highest_pagerank_node]
hamming_distances_lowest = [hamming(adj_matrix[node_to_index[lowest_pagerank_node], :], adj_matrix[:,node_to_index[node]]) for node in G_fb.nodes() if node != lowest_pagerank_node]
```

```python
# Plot the distributions of cosine similarity
# Plot the distributions of Jaccard similarity
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.hist(jaccard_values)
plt.title(f'Jaccard Similarity Distribution for Node {highest_pagerank_node}')
plt.xlabel('Similarity Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(cosine_similarities_highest)
plt.title(f'Cosine Similarity Distribution for Node {highest_pagerank_node}')
plt.xlabel('Similarity Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(hamming_distances_highest)
plt.title(f'Hamming Distance Distribution for Node {highest_pagerank_node}')
plt.xlabel('Distance Value')
plt.ylabel('Frequency')


plt.show()
```

```python
# Plot the distributions of cosine similarity, jacard similarity and Hamming distance for the node with the lowest PageRank value

plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.hist(jaccard_values_lowest)
plt.title(f'Jaccard Similarity Distribution for Node {lowest_pagerank_node}')
plt.xlabel('Similarity Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(cosine_similarities_lowest)
plt.title(f'Cosine Similarity Distribution for Node {lowest_pagerank_node}')
plt.xlabel('Similarity Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(hamming_distances_lowest)
plt.title(f'Hamming Distance Distribution for Node {lowest_pagerank_node}')
plt.xlabel('Distance Value')
plt.ylabel('Frequency')


plt.show()
```
